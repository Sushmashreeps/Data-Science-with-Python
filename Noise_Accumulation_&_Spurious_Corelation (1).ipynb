```python
import numpy as np
import matplotlib.pyplot as plt

# Generating synthetic data
np.random.seed(0)
x = np.arange(100)
y1 = np.sin(x/10) + np.random.normal(0, 0.2, 100)  # True signal with added noise
y2 = np.random.normal(0, 0.2, 100)  # Pure noise

# Adding noise accumulation
y1_accumulated = np.cumsum(y1)
y2_accumulated = np.cumsum(y2)

# Plotting
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y1, label='True Signal + Noise', color='blue')
plt.plot(x, y2, label='Pure Noise', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Noise Accumulation')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(y1_accumulated, y2_accumulated, color='green')
plt.xlabel('Signal Accumulated')
plt.ylabel('Noise Accumulated')
plt.title('Spurious Correlation')
plt.tight_layout()

plt.show()
```
