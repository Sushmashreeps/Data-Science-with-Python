```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(1,10)
x
```

```python
y = 2*x
y2 = 3*x
y2
```

```python
plt.plot(x,y, color='yellow', linewidth=5, linestyle=':')
plt.plot(x,y2, color='red', linewidth=5, linestyle=':')
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

Bar-plot

```python
student = {'sushma':100,'joyce':50,'anusha':20}
student
```

```python
names = list(student.keys())
marks = list(student.values())
```

```python
names, marks
```

```python
plt.bar(names, marks, color='pink')
plt.title("Marks of studend")
plt.xlabel("Names of Students")
plt.ylabel('Marks of students')
plt.grid(True)
plt.show()
```

Scatter plot

```python
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
```

```python
plt.scatter(x,y, color='pink')
plt.grid(True)
plt.show()
```

```python
fruits = ['apples',"banana",'oranges','mango']

cost = [95,40,70,200]

plt.pie(cost, labels=fruits, autopct='%.1f%%',shadow=True)
plt.figure(figsize=(1,1))
plt.show()
```
